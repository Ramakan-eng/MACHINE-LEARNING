**1.KMeans Clustering**

- find the optimal number of clusters:Elbow Curve

```python
wcss = []

for k in range(1,11):
    from sklearn.cluster import KMeans
    kmeans = KMeans(n_clusters = k, init = 'k-means++')
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)

print(wcss)
```
