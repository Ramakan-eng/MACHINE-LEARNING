```python
x = df.drop('expenses',axis = 1)
y = df['expenses']
```

```python
Train = []
cv = []

for i in range(1,100):
    from sklearn.model_selection import train_test_split 
    x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2,random_state = i)

    from sklearn.linear_model import ElasticNet
    model = ElasticNet()
    model.fit(x_train,y_train)

    ypred_train = model.predict(x_train)

    from sklearn.metrics import r2_score 
    Train.append(r2_score(y_train,ypred_train))
    from sklearn.model_selection import cross_val_score
    cv.append(cross_val_score(model,x_train,y_train,cv = 5).mean())

em = pd.DataFrame({'Train':Train,'cv':cv})
print(em)
gm = em[abs(em['Train'] - em['cv']) <= 0.5]
print(gm)
rs = gm[gm['cv'] == gm['cv'].max()].index.tolist()[0]
print('best random state number',rs)
```

### Step-4:Modelling & Evluation 

- Lasso Regression with default parameters

```python
# Modelling
from sklearn.linear_model import ElasticNet
enr_base = ElasticNet()
enr_base.fit(x_train,y_train)

# predictions 
train_predictions = enr_base.predict(x_train)
test_prediction = enr_base.predict(x_test)

#Evaluation 
print('Train R2:',enr_base.score(x_train,y_train))
print('Test R2:',enr_base.score(x_test,y_test))

from sklearn.model_selection import cross_val_score 
print('cross validation score:',cross_val_score(enr_base,x_train,y_train,cv=5).mean())
```

**enr_base.score(x_train, y_train)**

- and

- from sklearn.metrics import r2_score
- r2_score(y_train, enr_base.predict(x_train))
- Both give the same result (RÂ² score) â€” but the reason score() is used is for convenience. Let's dive in:

- ðŸ§  What is .score()?
- In scikit-learn, every model (like LinearRegression, Ridge, ElasticNet, etc.) has a .score() method.

- For regression models, .score() returns the RÂ² score:

- **model.score(X, y) == r2_score(y, model.predict(X))**

- So this line:

- **print('Train R2:', enr_base.score(x_train, y_train))**

- is just a shortcut for:

- from sklearn.metrics import r2_score

- **print('Train R2:', r2_score(y_train, enr_base.predict(x_train)))**

- âœ… Why use .score()?

- It's shorter and cleaner.

- Good for quick evaluation of regression models.

- You donâ€™t need to explicitly import r2_score.

- ðŸš« When not to use .score()?

- When you want to calculate other metrics like MAE, MSE, RMSE, etc.

- When you want more control (e.g., comparing training vs. test predictions with different metrics).

- If you're working with classification models, .score() returns accuracy, not RÂ².

- âœ… Summary:

|Method|	Returns|	Use case|
|------|-----------|------------|   
|.score(X, y)|	RÂ² (for regression)|	Quick model evaluation|
|r2_score(y, y_pred)	|RÂ²	|Custom metrics, more flexibility|

```python
from sklearn.model_selection import GridSearchCV

# model
estimator = ElasticNet()

# parameter & values
param_grid = {'alpha':[0.1,0.2,1,2,3,5,10],'l1_ratio':[0.1,0.5,0.75,0.9,0.95,1]}

#Identifying the best value of the parameter within given values for the given data 
model_hp = GridSearchCV(estimator,param_grid,cv = 5 ,scoring = 'neg_mean_squared_error')
model_hp.fit(x_train,y_train)
model_hp.best_params_
```

- ðŸ“¦ 1. Importing GridSearchCV

- from sklearn.model_selection import GridSearchCV

- GridSearchCV is a tool from scikit-learn used for hyperparameter tuning.

- It tries all combinations of parameters from a given grid and selects the best one based on a chosen evaluation metric.

- âš™ï¸ 2. Creating the model

- estimator = ElasticNet()

- ElasticNet is a regression model that combines L1 (Lasso) and L2 (Ridge) regularization.

- This line creates an instance of the model (with default settings) to be tuned.

- â„¹ï¸ You should import it beforehand:

- from sklearn.linear_model import ElasticNet

- ðŸ”¢ 3. Setting up the parameter grid

- param_grid = {
    'alpha': [0.1, 0.2, 1, 2, 3, 5, 10],
    'l1_ratio': [0.1, 0.5, 0.75, 0.9, 0.95, 1]
}
- This defines the values of hyperparameters you want to try:

- alpha: Controls overall regularization strength.

- Higher alpha â†’ stronger penalty â†’ simpler model (but might underfit).

- l1_ratio: Controls the mix between L1 and L2:

- l1_ratio = 1 â†’ pure Lasso,

- l1_ratio = 0 â†’ pure Ridge,

- Between 0 and 1 â†’ ElasticNet.

- This grid will try 7 Ã— 6 = 42 combinations.

- ðŸ” 4. Setting up GridSearchCV

- model_hp = GridSearchCV(estimator, param_grid, cv=5, scoring='neg_mean_squared_error')

- Letâ€™s explain each part:

- estimator: the base model to tune (ElasticNet here).

- param_grid: the dictionary of hyperparameter combinations to search over.

- cv=5: 5-fold cross-validation is used to evaluate each parameter combo.

- Data is split into 5 parts, trained on 4, tested on 1, and rotated.

- scoring='neg_mean_squared_error': the metric to evaluate each model.

- Uses negative MSE because scikit-learn expects a higher score to be better, so it flips the MSE (which is normally lower = better).

- ðŸƒâ€â™‚ï¸ 5. Fitting the model

- model_hp.fit(x_train, y_train)

- This actually runs the grid search over all parameter combinations using the training data.

- For each combination, it does cross-validation, computes the score, and finds the best set of parameters.

- ðŸ† 6. Getting the best parameters

- model_hp.best_params_

- This gives you the best combination of alpha and l1_ratio based on the cross-validation performance.

- âœ… Final Notes

- Term	Meaning

- ElasticNet()	Regularized regression model

- param_grid	Hyperparameter search space

- GridSearchCV	Searches the best combination of parameters
cv=5	5-fold cross-validation
scoring='neg_mean_squared_error'	Evaluation metric
- .fit()	Runs the full search
- .best_params_	Returns best combination of hyperparameters

**Rebuilt Lasso Model using best hyperparameters**

```python
# modelling 

enr_best = ElasticNet(alpha = 10,l1_ratio = 1)
enr_best.fit(x_train,y_train)

print('Intercept:',enr_best.intercept_)
print('coefficient:',enr_best.coef_)

# predictions 
train_predictions = enr_best.predict(x_train)
test_predictions = enr_best.predict(x_test)

# Evaluation 
print('Train R2:',enr_best.score(x_train,y_train))
print('Test R2:',enr_best.score(x_test,y_test))
print('cross validation score :', cross_val_score(enr_best,x_train,y_train,cv = 5).mean())
```
