```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.simplefilter('ignore')
```

```python
df = pd.read_csv(r'C:\Users\sahur\Downloads\DATASET OF DATA SCIENCE SRK SIR\mushrooms.csv')
df.head()
```

**X & y**

```python
input_data = df.drop('class',axis = 1)
X =pd.get_dummies(input_data,drop_first = True)
```

```python
y = pd.get_dummies(df['class'],drop_first = True)
```

**Train Test Split**

```python
Train = []
CV = []

for i in range(1,10):
    from sklearn.model_selection import train_test_split
    X_train,X_test,y_train,y_test = train_test_split(X,y ,test_size = 0.2 ,random_state = i)

    from xgboost import XGBClassifier
    model = XGBClassifier()
    model.fit(X_train,y_train)

    ypred_train = model.predict(X_train)
    ypred_test = model.predict(X_test)

    from sklearn.metrics import accuracy_score 
    Train.append(accuracy_score(y_train,ypred_train))
    
    from sklearn.model_selection import cross_val_score 
    CV.append(cross_val_score(model,X_train,y_train,cv = 5).mean())

    em = pd.DataFrame({'Train':Train,'CV':CV})
    gm = em[abs(em['Train']- em['CV'])<=0.5]
    rs = gm[gm['CV']==gm['CV'].max()].index.to_list()[0]

rs
```

```python
from sklearn.model_selection  import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 0)
```

**XGBoost Classifier Modes with default parameter**

```python
# modelling 
from xgboost import XGBClassifier 
xgb_model = XGBClassifier()
xgb_model.fit(X_train,y_train)
```

- ðŸ§  Line-by-Line Explanation
- 1ï¸âƒ£ Import the model

- from xgboost import XGBClassifier
- This imports the XGBoost Classifier from the XGBoost library. XGBoost stands for Extreme Gradient Boosting, which is an advanced implementation of gradient boosting machines (GBM).

- 2ï¸âƒ£ Create the model

- xgb_model = XGBClassifier()
- This initializes an XGBoost classifier with default parameters.

- Some default parameters include:

- n_estimators=100: number of trees

- learning_rate=0.3: how much each tree contributes

- max_depth=6: max depth of each tree

- booster='gbtree': type of model to run (tree-based boosting)

- ðŸ§  You can customize any of these to fine-tune performance.

- 3ï¸âƒ£ Train the model

- xgb_model.fit(X_train, y_train)
- This trains the XGBoost model on your training data.

- ðŸ”„ What happens during .fit()?
- Initialization: The model starts with a base prediction (like the average log-odds).

- Boosting rounds:

- Builds a sequence of decision trees.

- Each new tree learns from the errors (residuals) of the previous trees.

- The predictions are gradually improved (boosted).

- Gradient Descent: It uses gradients of a loss function (like log-loss for classification) to guide the direction of improvement.

- Regularization: Applies techniques to avoid overfitting (like L1/L2 penalties).

- âš™ï¸ Why Use XGBoost?

- ðŸ”¥ Feature	ðŸš€ Benefit
- Fast	Optimized in C++, parallelized boosting
- Regularization	Avoids overfitting better than normal GBM
- Handles missing data	No need to impute NaNs; it learns best splits using available data
- Tree pruning	Prunes trees using max_depth and min_child_weight
- Built-in CV & early stop	Can use built-in cross-validation and stop early if performance plateaus
- Feature importance	Built-in method to check which features are most predictive
- ðŸ“ˆ When Should You Use XGBoost?
- âœ… When:

- You have structured/tabular data

- You want high accuracy

- You're participating in competitions (e.g., Kaggle)

- Your dataset is clean and numeric (which youâ€™ve already preprocessed using pd.get_dummies())

- âŒ Avoid if:

- You're working on image, text, or audio data (Deep Learning is better here)

```python
ypred_train = xgb_model.predict(X_train)
ypred_test = xgb_model.predict(X_test)

from sklearn.metrics import accuracy_score
print('Train Accuracy:',accuracy_score(y_train,ypred_train))
print('Test Accuracy:',accuracy_score(y_test,ypred_test))

from sklearn.model_selection import cross_val_score 
print('Cross validation score :',cross_val_score(xgb_model,X_train,y_train,cv=5,scoring = 'accuracy').mean())
```

**Hyperparameter Tuning**

```python
from sklearn.model_selection import GridSearchCV
# model / estimators 
estimator = XGBClassifier()

# parameter grid
param_grid = {'n_estimators':[10,20,40,100] 
              , 'max_depth' : [3,4,5],
             'gamma':[0,0.15,0.3,0.5,1]}

# grid search 
grid = GridSearchCV(xgb_model,param_grid,cv = 5,scoring = 'accuracy')

# fit the data 
grid.fit(X_train,y_train)

# best parameters for XGB model
grid.best_params_
```

**Feature Importance**

```python
grid.best_estimator_.feature_importances_
```

```python
feats = pd.DataFrame(data = grid.best_estimator_.feature_importances_,
                    index = X.columns,
                    columns = ['Importance'])
important_features = feats[feats['Importance']>0.01]
imp_features_list = feats[feats['Importance']>0.01].index.to_list()
imp_features_list
```

**XGBoost with best hyperparameters**

```python
X_imp = X[imp_features_list]    # dropped unimportant 

X_train,X_test,y_train,y_test = train_test_split(X_imp,y,test_size = 0.2,random_state =101)
# modelling 
xgb_bhp = XGBClassifier(gamma = 0,max_depth = 3,n_estimators = 40)
xgb_bhp.fit(X_train,y_train)

# prediction 
ypred_train = xgb_bhp.predict(X_train)
prediction = xgb_bhp.predict(X_test)

# Evaluation 
from sklearn.metrics import accuracy_score
print('Train accuracy:',accuracy_score(y_train,ypred_train))
print('Test accuracy:',accuracy_score(y_test,prediction))

scores = cross_val_score(xgb_bhp,X_train,y_train,cv = 5)
print('Cross Validation Score',scores.mean())
```
